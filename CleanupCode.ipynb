```python
import os
import numpy as np
from PIL import Image
```

```python
# This is the code for moving LAG_dataset to child_PreGen_data 
path = 'LAGdataset_100/'
dest_folder = 'datasets/child_PreGen_data/'

for r, d, f in os.walk(path):
        all_children = d
        break

for child in all_children:
    child_path = path+child+'/y/'
    print(child_path)
    
    for r, d, f in os.walk(child_path):
        child_path_images = f
        break
    
    #move to src_img_path to dest_folder
    for child_path_image in child_path_images:
        src_img_path = child_path+child_path_image
        os.rename(src_img_path, dest_folder+child+'_'+child_path_image)
```

```python
child_folder = 'datasets/child_PreGen_data/'
for r, d, f in os.walk(child_folder):
    child_path_images = f
    break

for child_path in child_path_images:
    addr = child_folder+child_path
    print(addr)
    original_img = Image.open(addr).resize((64,64),Image.ANTIALIAS)
    original_img.save(addr)
```
